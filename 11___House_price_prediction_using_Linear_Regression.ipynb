# **Day-11 | House price prediction using Linear Regression-SingleVariable**

### *Import Libraries*

```python
import pandas as pd #loading data into python code using pandas library
from sklearn.linear_model import LinearRegression #loading an linear regression algo using sklearn library
import matplotlib.pyplot as plt # data visualization
```

### *Load Dataset from Local Directory*

```python
from google.colab import files
uploaded = files.upload()
```

### *Load Dataset*

```python
dataset = pd.read_csv('data2.csv') #load the data in the name dataset
```

### *Load Summarize*

```python
print(dataset.shape) #no. of rows and columns present in data
print(dataset.head(5)) #top 5 values of the data
```

### *Visualize Dataset*

```python
plt.xlabel('Level')
plt.ylabel('Salary')
plt.scatter(dataset.Level,dataset.Salary,color='red',marker='*')
```

### *Segregate Dataset into Input X & Output Y*

```python
X = dataset.drop('Salary',axis='columns')
X
```

```python
Y = dataset.Salary
Y
```

### *Training Dataset using Linear Regression*

```python
model = LinearRegression()   #training algorithm for our data
model.fit(X,Y)
```

### *Predicted Price for Land sq.Feet of custom values*

```python
x=18
LandAreainSqFt=[[x]]
PredictedmodelResult = model.predict(LandAreainSqFt)
print(PredictedmodelResult)
```

### Let's check is our model is Right ?
### Theory Calculation
### Y = m * X + b (m is coefficient and b is intercept)

*Coefficient - m*

```python
m=model.coef_
print(m)
```

*Intercept - b*

```python
b=model.intercept_
print(b)
```

### Y=mx+b
*x is Independant variable - Input - area*

```python
y = m*x + b
print("The Price of {0} Square feet Land is: {1}".format(x,y[0]))
```
